```python
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable


def sample_gumbel(shape, eps=1e-20):
    U = torch.rand(shape)
    return -Variable(torch.log(-torch.log(U + eps) + eps))


def gumbel_softmax_sample(logits, temperature):
    y = logits + sample_gumbel(logits.size())
    return F.softmax(y / temperature, dim=-1)


def gumbel_softmax(logits, temperature):
    """
    input: [*, n_class]
    return: [*, n_class] an one-hot vector
    """
    y = gumbel_softmax_sample(logits, temperature)
    shape = y.size()
    _, ind = y.max(dim=-1)
    y_hard = torch.zeros_like(y).view(-1, shape[-1])
    y_hard.scatter_(1, ind.view(-1, 1), 1)
    y_hard = y_hard.view(*shape)
    return (y_hard - y).detach() + y
```

```python
def calculate_sampled_distribution_gumble_softmax(x_attention) -> torch.Tensor:
    """
    :return binary tensor of sampled by gumble-softmax with size of [n_patches + 1] # 1 for [CLS] token
    When X_attention value is large, the sample should be 1. The first token is for [CLS] and should be always turn-on.
    """
    sigmoid_x_attention = torch.sigmoid(x_attention).clone()
    log_prob_x_attention = torch.stack((torch.log(1 - sigmoid_x_attention), torch.log(sigmoid_x_attention))).T
    log_probs = [gumbel_softmax(log_prob, temperature=1) for log_prob in
                 log_prob_x_attention]
    sampled_binary_patches = torch.stack(log_probs) @ torch.tensor([0., 1.]).T
    sampled_binary_patches[0] = torch.tensor(1.0)  # For [CLS] token
    return sampled_binary_patches
```

```python
def calculate_sampled_distribution_gumble_softmax(x_attention) -> torch.Tensor:
    """
    :return binary tensor of sampled by gumble-softmax with size of [n_heads, n_patches + 1] # 1 for [CLS] token
    When X_attention value is large, the sample should be 1. The first token is for [CLS] token and should be turn-on.
    """
    n_heads, n_tokens = x_attention.shape

    sigmoid_x_attention = torch.sigmoid(x_attention).clone()
    log_prob_x_attention = torch.stack(
        (torch.log(1 - sigmoid_x_attention), torch.log(sigmoid_x_attention))).reshape(n_heads, n_tokens, 2)
    # print(log_prob_x_attention.shape)
    log_probs = [gumbel_softmax(p_dist, temperature=1) for log_prob in log_prob_x_attention for p_dist in log_prob]
    sampled_binary_patches = (torch.stack(log_probs) @ torch.tensor([0., 1.]).T).reshape(n_heads, n_tokens)
    sampled_binary_patches[:, 0] = torch.tensor(1.0)
    return sampled_binary_patches
```
